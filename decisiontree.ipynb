```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz,plot_tree
from IPython.display import Image
```

```python
df=pd.read_csv("/content/drive/MyDrive/DATASET/demoDT.csv",sep=",")
df
```

```python
features_cols=['Literacy',' Cleanliness','CRIME']
feature=df[features_cols]
Target=df.GOOD
```

```python
DT = DecisionTreeClassifier()
DT = DT.fit(feature,Target)
```

```python
l=int(input("Enter Literacy: "))
cl=int(input("Enter cleanliness: "))
cr=int(input("Enter crime rate: "))
pred=DT.predict([[l,cl,cr]])
if pred==1:
  print("Good State")
else:
  print("Bad state")
```

```python
plot_tree(DT)
```
