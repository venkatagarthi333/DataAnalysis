```python
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset("tips")
print(tips)
sns.scatterplot(x='total_bill',y='tip',data=tips,label="tips")

plt.title("scatter plot of total bill vs trip")
plt.xlabel("total bill ($)")
plt.ylabel("tip ($)")
plt.show()
```

```python
iris=sns.load_dataset("iris")
print(iris)
sns.scatterplot(x='sepal_length',y='sepal_width',data=iris)

plt.title("plant details")
plt.xlabel("sepal_length ($)")
plt.ylabel("sepal_width($)")
```

```python
print(tips)
sns.barplot(x='total_bill',y='tip',data=tips,label="tips")

plt.title("bar plot of total bill  by day")
plt.xlabel("total bill ($)")
plt.ylabel("day ($)")
plt.show()
```

```python
print(tips)
sns.boxplot(x='day',y='total_bill',data=tips)

plt.title("Box plot of total bill vs trip")
plt.xlabel("day ($)")
plt.ylabel("total_bill ($)")
plt.show()
```

```python
print(tips)
sns.lineplot(x='day',y='total_bill',data=tips)

plt.title("violin plot of total bill vs trip")
plt.xlabel("day ($)")
plt.ylabel("total_bill ($)")
plt.show()
```
